```python
# Import Json files
import json
path = 'simulation_1.json'
with open(path, 'r') as f:
    simulation = json.load(f)
```

```python
# Format Output:
from termcolor import colored, cprint

for post in simulation:
    post_content = f'''\n\n**************************************************************************************
{post['creator']} wrote this post:
TITLE: {post['title']}
CONTENT: {post['content']}

It received {len(post['comments'])} comments:'''
    cprint(post_content)
    comments = ''
    for i,comment in enumerate(post['comments']):
        string = f'{comment[1]}'
        comments += '\n\n'+string
    print(colored(comments,'blue'))
```
